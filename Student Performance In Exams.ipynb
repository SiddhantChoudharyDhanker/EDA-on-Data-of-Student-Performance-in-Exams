```python
df['race/ethnicity'].value_counts()
```

```python
import matplotlib.pyplot as plt
```

```python
race_ethnicity = df['race/ethnicity'].value_counts()
plt.pie(x = race_ethnicity.values, labels = race_ethnicity.index, autopct='%1.1f%%', shadow=True, startangle=90)
plt.title('Race/Ethnicity distribution')
plt.show()
```

- Group C and Group D account for a larger proportion that other groups, Group A has at least 8.9%.

```python
level = df['parental level of education'].value_counts()
```

```python
level
```

```python
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i, y[i], y[i], ha = 'center')


fig = plt.figure(figsize = (10,5))
plt.bar(list(level.index), list(level.values), width = 0.4)
plt.xlabel("Level of Education")
plt.ylabel("Count")
plt.title("Parental level of education distribution")
addlabels(list(level.index), list(level.values))
plt.show()
```

- The majority of parents have studied some college and associate's degree, the number of master's degree accounted for less.

### Numeric Columns

```python
df.head()
```

```python
plt.figure(figsize = (18,8))
plt.subplot(1,4,1)
plt.title('MATH SCORE')
sns.violinplot(y = 'math score', data = df, color = 'red', linewidth = 3)
plt.subplot(1,4,2)
plt.title('READING SCORE')
sns.violinplot(y = 'reading score', data = df, color = 'green', linewidth = 3)
plt.subplot(1,4,3)
plt.title('WRITING SCORE')
sns.violinplot(y = 'writing score', data = df, color = 'blue', linewidth = 3)
```

- From the above three plots its clearly visible that most of the student score in between 60-90 in Maths whereas in reading and writing most of them score from 50-80.

```python
math_score = pd.to_numeric(df['math score'], downcast='float')
reading_score = pd.to_numeric(df['reading score'], downcast='float')
writing_score = pd.to_numeric(df['writing score'], downcast='float')
```

```python
df.drop(df.iloc[:, 8:], inplace = True, axis = 1)
```

```python
print("Average math score is : {}".format(np.mean(math_score)))
print("Average reading score is : {}".format(np.mean(reading_score)))
print("Average writing score is : {}".format(np.mean(writing_score)))
print("Average total score is : {}".format(np.mean(math_score + reading_score + writing_score)/3))
```

- Performance by students in Math section is lower when compared to reading and writing
- Best Performance is in reading section

# Multivariate Analysis

```python
fig = sns.pairplot(df[['math score', 'reading score', 'writing score', 'gender']], hue = 'gender', height = 5)
fig.fig.suptitle("How are the different scores related?", size=30, y=1.05)
```

```python
plt.figure(figsize = (16, 9))
sns.heatmap(df.corr(), annot=True, square = True)
plt.title("Confusion Matrix", fontsize = 20, fontstyle = "oblique")
```

- From the heatmap, it can be seen that between the types of scores have a </mark>correlation relationship with each other</mark>. Specifically, this high score tends to lead to other types of scores as well. 

#### Score and Gender

To check the trend between gender and score, I will calculate the average score by gender.

```python
gender_mean = df.groupby('gender')[['math score', 'reading score', 'writing score']].mean().round(2).T
gender_mean
```

- From the table above, it can be seen that, the average score of math of the male is higher than that of the female, while the average score of reading and writing of the female is higher than that of the male.
